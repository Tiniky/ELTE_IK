```python
# írjunk föggvényt amely adott értékig megszámolja a primek számát
def db_prim(n):
    if n < 2:
        print("nincs prim")
    p = 2
    s = 0
    while p < n:
        s += 1
        p = next_prime(p)
    return s

print(db_prim(100))
```
