```python
# ezek segitségével írd meg a skaláris szorzat függvényét (hf)
def skalar(L1, L2):
    L = []
    for elem in L1:
        L += zip([elem]*len(L2), L2)
    return L

test1 = [1,2,3]
test2 = ['a','b','c']
skalar(test1, test2)
```

```python
# írjunk föggvényt amely adott értékig megszámolja a primek számát
def db_prim(n):
    if n < 2:
        print("nincs prim")
    p = 2
    s = 0
    while p < n:
        s += 1
        p = next_prime(p)
    return s

print(db_prim(100))
```

```python
# reflexiv minden a: (a,a)
def reflexiv(L):
    refl = True
    for tmp in L:
        if(tmp[0] != tmp[1]):
            refl = False
            break
    return refl

tt = [[1,2], [1,1], [1,2]]
tt2 = [[1,1], [2,2]]

print(reflexiv(tt))
print(reflexiv(tt2))
```

```python
# tranzitiv (a, b), (b, c) -> (a,c)
def tranzitiv(L):
    tranz = True
    for tmp in L:
        for tmp2 in L:
            if(tmp[1] == tmp2[0]):
                for tmp3 in L:
                    if(tmp[0] == tmp3[0] and tmp2[1] == tmp3[1]):
                        tranz = True
                    else:
                        tranz = False
                        break
    return tranz

tt = [[1,2], [2,3], [1,3]]
tt2 = [[1,1], [2,2]]

print(tranzitiv(tt))
print(tranzitiv(tt2))
```
